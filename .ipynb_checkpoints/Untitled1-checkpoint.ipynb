```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_log_error
from sklearn.ensemble import GradientBoostingRegressor
```

```python
import warnings
warnings.filterwarnings("ignore")

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("train.csv")
```

```python
df.columns = df.columns.str.lower().str.strip()
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
```

```python
df['ship_date'] = pd.to_datetime(df['ship_date'], dayfirst=True)
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['day'] = df['order_date'].dt.day
df['dayofweek'] = df['order_date'].dt.dayofweek
df['quarter'] = df['order_date'].dt.quarter
```

```python
df['shipping_days'] = (df['ship_date'] - df['order_date']).dt.days
```

```python
drop_cols =[
    'row_id', 'order_id', 'customer_id',
    'customer_name', 'product_id',
    'ship_date', 'ship_mode', 
    'postal_code', 'region'
]

df = df.drop(columns=drop_cols, errors='ignore')
```

```python
df.head()
```

```python
monthly_sales = df.groupby('order_date')['sales'].sum()
```

```python
monthly_sales = (
    df.groupby([
        pd.Grouper(key='order_date', freq='M'),
        'state',
        'sub-category'
    ])['sales']
    .sum()
    .reset_index()
)
```

```python
monthly_sales.head()
```

```python
monthly_sales = monthly_sales.sort_values(['state', 'sub-category', 'order_date',])

monthly_sales['lag_1'] = (
    monthly_sales
    .groupby(['state', 'sub-category'])['sales']
    .shift(1)
)

monthly_sales['lag_2'] = (
    monthly_sales
    .groupby(['state', 'sub-category'])['sales']
    .shift(2)
)


monthly_sales['rolling_3'] = (
    monthly_sales
    .groupby(['state', 'sub-category'])['sales']
    .transform(lambda x: x.rolling(3).mean())
)
```

```python
monthly_sales.head(20)
```

```python
monthly_sales['time_index'] = (
    (monthly_sales['order_date'] - monthly_sales['order_date'].min()).dt.days
)
```

```python
monthly_sales['month'] = monthly_sales['order_date'].dt.month
```

```python
monthly_sales['month_sin'] = np.sin(2 * np.pi * monthly_sales['month']/12)
monthly_sales['month_cos'] = np.cos(2 * np.pi * monthly_sales['month']/12)
```

```python
monthly_sales['momentum_1'] = (
    monthly_sales['lag_1'] - monthly_sales['lag_2']
)
```

```python
monthly_sales['subcategory_avg'] = (
    monthly_sales
    .groupby(['sub-category', 'order_date'])['sales']
    .transform('mean')
)
```

```python
monthly_sales = monthly_sales.dropna()
```

```python
monthly_sales.head()
```

```python
train = monthly_sales[monthly_sales['order_date'] < '2018-01-01']
test = monthly_sales[monthly_sales['order_date'] >= '2018-01-01']
```

```python
target = 'sales'

categorical_cols = ['sub-category', 'state']
numerical_cols = [
    'lag_1', 'lag_2', 'rolling_3', 'time_index',
     'month_sin', 'month_cos', 'momentum_1', 'subcategory_avg'
]

X_train = train[categorical_cols + numerical_cols]
y_train = train[target]

X_test = test[categorical_cols + numerical_cols]
y_test = test[target]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', 'passthrough', numerical_cols)
    ]
)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
!pip install catboost
```

model = RandomForestRegressor(
    n_estimators = 200,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

```python

from catboost import CatBoostRegressor

model = CatBoostRegressor(
    iterations=1000,
    depth=6,
    learning_rate=0.05,
    l2_leaf_reg=5,
    random_state=42,
    verbose=200
)
```

```python
pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ]
)
```

```python
pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

preds = pipeline.predict(X_test)

print("MAE:", mean_absolute_error(y_test, preds))
print("R2:", r2_score(y_test, preds))
```

```python
preds = np.maximum(preds, 0)
```

```python
from sklearn.metrics import mean_squared_log_error
rmsle = np.sqrt(mean_squared_log_error(y_test, preds))
print("RMSLE:", rmsle)
```
